## what is TensorFlow?
Tensorflow is a powerful open-source end-to-end software library developed by Google Brain Team in 2015 for deep neural networks. Tensorflow is so popular for the following reasons:
- It works with all popular languages such as python, c++, Java, R and GO
- It uses Keras as a high-level API - it specifies how software component should interact
- Tensorflow alllows model deployment and ease of use in production
- Tensorflow has a very good online community with over 5000 contributors on github

## What is Keras?
Keras is an open-source library that provides a Python interface for artificial neural networks. Keras acts as an interface for the TensorFlow library. In TensorFlow 2.0, Keras has been adopted as the standard high-level API. Keras can be integraed with multiple deep learning engines including Google TensorFlow, Amazon MXNET, Micorsoft CNTK and Theano

Before we continue to learn the basic concepts of TensorFlow and Keras, we will first build our own deep learning model with TensorFlow


```python
import tensorflow as tf
from tensorflow import keras
import numpy as np

#network and training parameters
EPOCHS = 200
BATCH_SIZE = 128
VERBOSE = 1
NB_CLASSES  = 10 #number of output - 10 digits
N_HIDDEN = 128
VALIDATION_SPLIT = 0.2

#labels have one-hot representation is automatically applied
mnist = keras.datasets.mnist
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

#X_train is 60000 rows of 28x28 values so we reshape to 6000 x 784
RESHAPED = 784

X_train = X_train.reshape(60000, RESHAPED)
X_test = X_test.reshape(10000, RESHAPED)

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

#normalize inputs to be writtent in [0,1]
X_train = X_train/255
X_test = X_test/255
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')

#one-hot encoding representation of the lables
Y_train = tf.keras.utils.to_categorical(Y_train, NB_CLASSES)
Y_test = tf.keras.utils.to_categorical(Y_test, NB_CLASSES)
#buidl your model
model = tf.keras.models.Sequential()
model.add(keras.layers.Dense(NB_CLASSES, input_shape=(RESHAPED, ), kernel_initializer='zeros', name='dense_layer', activation='softmax'))
```

```python
model.summary()
```
